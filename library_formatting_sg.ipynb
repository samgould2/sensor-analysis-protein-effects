```python
import pandas as pd

# Function to calculate protospacer coordinates
def calculate_protospacer_coordinates(row):
    mutation_to_pam_distance = row['Mut_PAM_Distance']
    mutation_position = row['Start_Position']
    
    if mutation_to_pam_distance >= 0:  # Positive strand
        pam_position = mutation_position + mutation_to_pam_distance
        protospacer_end = pam_position - 1
        protospacer_start = protospacer_end - 19
    else:  # Negative strand
        pam_position = mutation_position + mutation_to_pam_distance
        protospacer_end = pam_position + 1
        protospacer_start = protospacer_end + 19

    return pd.Series({'Protospacer_Start': protospacer_start, 'Protospacer_End': protospacer_end})

lib[['Protospacer_Start', 'Protospacer_End']] = lib.apply(calculate_protospacer_coordinates, axis=1)

lib
```

```python
import Bio.Seq
import pandas as pd
from pegg import prime

#change this to your filepath for your reference genome
filepath = '/Users/diegodetres/Downloads/GCF_000001405.25_GRCh37.p13_genomic.fna.gz'
chrom_dict, i = prime.genome_loader(filepath)
#chrom_dict = genome sequences
```

```python
chrom_dict.keys()
```

```python
t = lib
targ = t[t['type_of_guide']=='targeting']
t_small = targ[['Protospacer_Matched', 'Chromosome', 'Hugo_Symbol', 'Protospacer_Start', 'Protospacer_End', 'PAM_Strand']]

def chromosome_to_key(chrom):
    if chrom in ['X', 'Y']:
        return chrom  
    else:
        return int(chrom)

for i, val in t_small.iterrows():
    s = int(val['Protospacer_Start'])
    e = int(val['Protospacer_End'])
    c = chromosome_to_key(val['Chromosome'])
    strand = val['PAM_Strand']


    protospacer = val['Protospacer_Matched']

    if strand =='+':
        proto = chrom_dict[c][s-1:e].upper()
        assert proto == protospacer, print(proto, protospacer)

    elif strand == '-':
        #in this case, proto_start is LARGER than proto_end
        proto = Bio.Seq.Seq(chrom_dict[c][e-1:s].upper()).reverse_complement()
        assert proto == protospacer, f"Assertion failed: proto ({proto}) != protospacer ({protospacer})"
```
